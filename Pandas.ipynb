```python
import pandas as pd
import numpy as np
```

```python
ao = pd.read_csv("D:\\SelfLearning\\Py\\Pandas\\AuditOutput.csv")

# list(ao) # to get header
sample_ao = ao.head(100) # to get top 5 rows
sample_ao
```

```python
##list all the columns which are not being computed, i.e. getting only one values in freq.
removeCols = list([col for col in list(sample_ao) if(len(sample_ao[col].value_counts().to_dict()) <= 1)])

## remove all the columns which are not being computed, i.e. getting only one values in freq.
# sample_ao = sample_ao.drop([col for col in removeCols], 1)
# sample_ao
# removeCols.remove('I_CumChargeAcctBalOver6Mos_ ')
ignoreCols = ['O_Scorecard_', 'O_ErrorCode_', 'O_OnlineBatchIndicator_', 'O_Tag_']
[removeCols.remove(x) for x in ignoreCols]
removeCols
```

```python
sample_ao = sample_ao.drop([col for col in removeCols], 1) ## drop columns
```

```python
list(sample_ao) ## print headers 
```

```python
sample_ao.sort_values("Sequence", inplace=True) ## sort dataframe
sample_ao
```

```python
## creating a series which contains conditions for where clause.
filter0 = sample_ao['I_gPmt1to6ByAmt1to6_'] > 0
filter1 = sample_ao['I_gAvgCallTime1to3Mby4to6M_'] > 0
## multi condition in where clause
sample_ao1 = sample_ao.where(filter0 &  filter1, inplace=False, other=np.nan)
## remove NaN
sample_ao1 = sample_ao1[np.isfinite(sample_ao1['I_gPmt1to6ByAmt1to6_'])]
sample_ao1
```

```python
# sample_ao2 = {'I_gMaxAmtOverdue18M_' : sample_ao1['I_gMaxAmtOverdue18M_'], 'O_Field01_' : sample_ao1['O_Field01_']}
sample_ao2 = pd.DataFrame(data={'Sequence' : ao['Sequence'],'I_gMaxAmtOverdue18M_' : ao['I_gMaxAmtOverdue18M_'], 'O_Field01_' : ao['O_Field01_']})
filter = sample_ao2['I_gMaxAmtOverdue18M_'] != sample_ao2['O_Field01_']
sample_ao2 = sample_ao2.where(filter, inplace=False, other=np.nan)
sample_ao2 = sample_ao2[np.isfinite(sample_ao2['I_gMaxAmtOverdue18M_'])]
sample_ao2
```
